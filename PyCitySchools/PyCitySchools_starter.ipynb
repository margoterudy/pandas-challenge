## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create a pandas series for each grade w conditional statements
math_nineth = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
math_tenth = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
math_eleventh = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
math_twelveth = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

#create the Dataframe
math_avg = pd.DataFrame({
        "9th": math_nineth,
        "10th": math_tenth,
        "11th": math_eleventh,
        "12th": math_twelveth
})

math_avg = math_avg[['9th', '10th', '11th', '12th']]
math_avg.index.name = "School"

math_avg
```

```python
#Format
math_avg.style.format({'9th': '{:.1f}', 
                          "10th": '{:.1f}', 
                          "11th": "{:.1f}", 
                          "12th": "{:.1f}"})
```
